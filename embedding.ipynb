```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
# sentences
sent_list = ['How are you?', 'My name is Vansh', 'I like coffee' , 'Go to the gym coward', 'the glass of milk', 'the glass of juice', 'she is very beautiful']
```

```python
# vocabulary size
voc_size =10000
```

```python
# one hot representation

one_hot_repr = [ one_hot(words, voc_size) for words in sent_list]
one_hot_repr
```

```python
# word embedding representation

from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding
import numpy as np
```

```python
sent_len = 8
embedded_docs = pad_sequences(one_hot_repr, padding='pre', maxlen=sent_len)
print(embedded_docs)
```

```python
# feature representation

dim = 10
```

```python
model = Sequential()

# Add embedding layer
model.add(Embedding(input_dim=voc_size, output_dim=dim, input_length=sent_len))

# Compile the model
model.compile(optimizer='adam', loss='mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
